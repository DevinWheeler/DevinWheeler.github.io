```python
from AnimalShelter import AnimalShelter

# Instantiate the AnimalShelter class
shelter = AnimalShelter()

# Define a sample animal to insert
sample_insert = {
    "age_upon_outcome": "2 years",
    "animal_id": "A720438",
    "animal_type": "Dog",
    "breed": "Pitbull",
    "color": "Black",
    "date_of_birth": "2023-07-14",
    "datetime": "2024-05-20 09:22:00",
    "monthyear": "2024-05",
    "name": "Spanky",
    "outcome_subtype": "",
    "outcome_type": "Adoption",
    "sex_upon_outcome": "Neutered Male",
    "location_lat": 30.7137,
    "location_long": -97.4682
}

# Test the create method
print("Inserting document:")
insert_result = shelter.create(sample_insert)
print(f"Insert successful: {insert_result}")

# Check for create
print("Checking for inserted document:")
create = shelter.read({"name": "Spanky"})
for animal in create:
    print(animal)

# Failing test case for create method
print("\nInserting invalid data:")
try:
    insert_result = shelter.create(None)
    print(f"Insert successful: {insert_result}")
except ValueError as e:
    print(f"Insert failed: {e}")

# Define a query to read documents
query = {"animal_type": "Dog"}

# Test the read method
print("\nQuerying documents:")
read_result = shelter.read(query)
print(f"Documents found: {len(read_result)}")

# Failing test case for read method
print("\nQuerying documents with no matches:")
read_result = shelter.read({"animal_type": "Turtles"})
print(f"Documents found: {len(read_result)}")

# Define an update query and update data
update_query = {"animal_id": "A720438"}
update_data = {"color": "Brown"}

# Test the update method
print("\nUpdating document:")
update_result = shelter.update(update_query, update_data)
print(f"Documents updated: {update_result}")

# Check for update
print("Checking for updated document:")
updated_document = shelter.read(update_query)
for animal in updated_document:
    print(animal)

# Failing test case for update method
print("\nUpdating with invalid data:")
try:
    update_result = shelter.update(None, update_data)
    print(f"Update successful: {update_result}")
except ValueError as e:
    print(f"Update failed: {e}")

# Define a delete query
delete_query = {"animal_id": "A720438"}

# Test the delete method
print("\nDeleting document:")
delete_result = shelter.delete(delete_query)
print(f"Documents deleted: {delete_result}")

# Check for delete
print("Checking for deleted document:")
deleted_document = shelter.read(delete_query)
print(f"Documents found after delete: {len(deleted_document)}")

# Failing test case for delete method
print("\nDeleting with invalid query:")
try:
    delete_result = shelter.delete(None)
    print(f"Delete successful: {delete_result}")
except ValueError as e:
    print(f"Delete failed: {e}")
```
